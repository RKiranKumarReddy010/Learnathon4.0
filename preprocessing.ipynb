```python
# Importing the required dependencies
import pandas as pd
from utils._datecleaner import DateTimeCleaner
import numpy as np
from utils._preprocessor import DataPreprocessor
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
train_data = pd.read_csv("data/Auto_Insurance_Fraud_Claims_File01.csv")
validation_data = pd.read_csv("data/Auto_Insurance_Fraud_Claims_File02.csv")
test_data = pd.read_csv("data/Auto_Insurance_Fraud_Claims_File03.csv")
```

***Preprocess the Data***

```python
# Manage the Date time data format
a = ["Bind_Date1","Policy_Start_Date","Policy_Expiry_Date","Accident_Date","DL_Expiry_Date","Claims_Date"]
for i in a:
    train_data[i] = train_data[i].apply(lambda x: DateTimeCleaner.parse_mixed_date(x) if pd.notna(x) else x)
    validation_data[i] = validation_data[i].apply(lambda x: DateTimeCleaner.parse_mixed_date(x) if pd.notna(x) else x)
    test_data[i] = test_data[i].apply(lambda x: DateTimeCleaner.parse_mixed_date(x) if pd.notna(x) else x)
```

```python
train_data.isnull().sum()
```

```python
def assignment_Policy_BI(data, column:str, arg1:str, arg2:str ):
    split_data = DataPreprocessor.split_waiting_indemnity(data[column])
    data[arg1] = [x[0] for x in split_data]
    data[arg2] = [x[1] for x in split_data]

#calling the function
assignment_Policy_BI(train_data, "Policy_BI","Wait_Policy_BI","Indemnity_Policy_BI")
assignment_Policy_BI(validation_data, "Policy_BI","Wait_Policy_BI","Indemnity_Policy_BI")
assignment_Policy_BI(test_data, "Policy_BI","Wait_Policy_BI","Indemnity_Policy_BI")
```

```python
@staticmethod
def divide_columns_by_index(data, col1_idx, col2_idx, new_col_name):
    col1 = data.columns[col1_idx]
    col2 = data.columns[col2_idx]
    data[new_col_name] = data[col1].astype(float) / data[col2].astype(float)
divide_columns_by_index(train_data,47,41,"Claim_Intensity")
divide_columns_by_index(validation_data,47,41,"Claim_Intensity")
divide_columns_by_index(test_data,47,41,"Claim_Intensity")
```

```python
def tranformed(data, column):
    enc = LabelEncoder()
    data[column] = enc.fit_transform(data[column])

#TRAIN DATA
tranformed(train_data, "Gender")
tranformed(train_data, "Accident_Type")
tranformed(train_data, "Collision_Type")
tranformed(train_data, "authorities_contacted")
tranformed(train_data, "Accident_Severity")
tranformed(train_data, "Property_Damage")
tranformed(train_data, "Police_Report")
tranformed(train_data, "Auto_Make")
tranformed(train_data, "Auto_Model")
tranformed(train_data, "Auto_Year")
tranformed(train_data, "Vehicle_Color")
tranformed(train_data, "Vehicle_Cost")
tranformed(train_data, "Fraud_Ind")

#TEST DATA
tranformed(test_data, "Gender")
tranformed(test_data, "Accident_Type")
tranformed(test_data, "Collision_Type")
tranformed(test_data, "authorities_contacted")
tranformed(test_data, "Accident_Severity")
tranformed(test_data, "Property_Damage")
tranformed(test_data, "Police_Report")
tranformed(test_data, "Auto_Make")
tranformed(test_data, "Auto_Model")
tranformed(test_data, "Auto_Year")
tranformed(test_data, "Vehicle_Color")
tranformed(test_data, "Vehicle_Cost")

#VALIDATE DATA
tranformed(validation_data, "Gender")
tranformed(validation_data, "Accident_Type")
tranformed(validation_data, "Collision_Type")
tranformed(validation_data, "authorities_contacted")
tranformed(validation_data, "Accident_Severity")
tranformed(validation_data, "Property_Damage")
tranformed(validation_data, "Police_Report")
tranformed(validation_data, "Auto_Make")
tranformed(validation_data, "Auto_Model")
tranformed(validation_data, "Auto_Year")
tranformed(validation_data, "Vehicle_Color")
tranformed(validation_data, "Vehicle_Cost")
tranformed(validation_data, "Fraud_Ind")
```
