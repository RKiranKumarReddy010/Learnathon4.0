```python
# Create a correlation matrix for numerical columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data[numeric_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Numerical Features')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

key_features = ['Age_Insured', 'Customer_Life_Value1', 'Policy_Premium', 'Vehicle_Claim', 'Property_Claim', 'Claim_Intensity']
plt.figure(figsize=(12, 8))
sns.pairplot(data[key_features], diag_kind='kde')
plt.suptitle('Pairwise Relationships Between Key Features', y=1.02)
plt.show()

features_for_parallel = ['Age_Insured', 'Customer_Life_Value1', 'Policy_Premium', 'Vehicle_Claim', 'Property_Claim']
parallel_data = data[features_for_parallel].copy()

for column in parallel_data.columns:
    parallel_data[column] = (parallel_data[column] - parallel_data[column].min()) / (parallel_data[column].max() - parallel_data[column].min())

parallel_data['Fraud_Ind'] = data['Fraud_Ind']

plt.figure(figsize=(12, 6))
parallel_coordinates(parallel_data, 'Fraud_Ind', colormap=plt.cm.get_cmap("Set2"))
plt.title('Parallel Coordinates Plot of Key Features')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot distribution of Claim_Intensity
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Claim_Intensity', kde=True, stat="density")
plt.title('Distribution of Claim Intensity')
plt.xlabel('Claim Intensity')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
